```python
from tabula import read_pdf

```

```python
file_name = "../acc.pdf"
```

```python
df = read_pdf("../acc.pdf")
```

```python
df
```

```python
# set pages option
read_pdf("../acc.pdf", pages=3)
```

```python
# pass pages as string
read_pdf("../acc.pdf", pages="1-2,3")
```

```python
# extract all pages
read_pdf("../acc.pdf", pages="all")
```

```python
# set area option
read_pdf("../acc.pdf", area=(126,149,212,462), pages=2)
```

```python
# read pdf as JSON
read_pdf("data.pdf", output_format="json")
```

```python
# You can convert from pdf into JSON, CSV, TSV
from tabula import convert_into

convert_into("../acc.pdf", "test.json", output_format="json")
!cat test.json
```

```python
convert_into("../acc.pdf", "test.tsv", output_format="tsv")
!cat test.tsv
```

```python
convert_into("../acc.pdf", "test.csv", output_format="csv")
!cat test.csv
```
